```
import matplotlib.pyplot as plt
import numpy as np
import time as t

def isValid(sudoku, cell, digit):
  
  row = cell // n
  col = cell % n
  cuadrant_row = (row // sqrt_n) * sqrt_n
  cuadrant_col = (col // sqrt_n) * sqrt_n


  for i in range(n):
    if (i != col) and (sudoku[row][i]== digit):
      return False
  for i in range(n):
    if (i != row) and (sudoku[i][col]== digit):
      return False
  for i in range(cuadrant_row, cuadrant_row+sqrt_n):
    for j in range(cuadrant_col, cuadrant_col+sqrt_n):
      if (i!=row and j!=col) and (sudoku[i][j]==digit):
        #print(f"i:{i}; j:{j}; digit:{digit}")
        #print(np.matrix(sudoku))
        return False
  return True


def sudoku_helper(sudoku, cell):
  if cell == n**2:
    return True
  #print(f"{cell}:{row}, {col}")
  for digit in range(1,n+1):
    row = cell // n
    col = cell % n  
    if(isValid(sudoku,cell, digit)): 
      sudoku[row][col] = digit
      if sudoku_helper(sudoku, cell+1):
        return True
      else:
        sudoku[row][col] = 0
  return False
    
def sudoku_solver(sudoku):
  sudoku_helper(sudoku, 0)
  print(np.matrix(sudoku))

times = {}
n=4
sqrt_n = int(n**0.5)
start_time = t.time()
sudoku = [[0 for col in range(n)] for row in range(n)]
sudoku[n-1][n-1] = 1
sudoku_solver(sudoku)
times[n]= t.time() - start_time
print("--- %s seconds ---" % (times[n]))

n=9
sqrt_n = int(n**0.5)
start_time = t.time()
sudoku = [[0 for col in range(n)] for row in range(n)]
sudoku[n-1][n-1] = int(n/2)
sudoku_solver(sudoku)
times[n]= t.time() - start_time
print("--- %s seconds ---" % (times[n]))

n=16
sqrt_n = int(n**0.5)
start_time = t.time()

sudoku = [[0 for col in range(n)] for row in range(n)]

sudoku[n-1][n-1] = int(n/2)
sudoku_solver(sudoku)
times[n]= t.time() - start_time
print("--- %s seconds ---" % (times[n]))



print(times)

lists = sorted(times.items())
x, y = zip(*lists)
plt.plot(x, y)

plt.show()
```
