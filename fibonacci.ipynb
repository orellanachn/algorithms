Fibonacci recursivo sin memorizaciÃ³n

```
def fibonacci(index):
  if index == 0:
    return 0
  elif index == 1 or index== 2:
    return 1
  return fibonacci(index-1)+ fibonacci(index-2)

print(fibonacci(7))
```

Fibonacci basado de Dynamic Programming


```
def fibonacci_helper(fibonacci_array, index):
  if fibonacci_array[index]>=0:
    return fibonacci_array[index]
  if index == 0:
    return 0
  elif index == 1 or index== 2:
    return 1
  fibonacci_array[index] = fibonacci_helper(fibonacci_array,index - 2) + fibonacci_helper(fibonacci_array, index - 1)
  return fibonacci_array[index]

def fibonacci_DP(index):
  fibonacci_array = [-1]*(index+1)
  return fibonacci_helper(fibonacci_array, index)

print(fibonacci_DP(7))
```

```
import matplotlib.pyplot as plt
import numpy as np
import time as t

times = {}
times_DP = {}

def test(n):
  start_time = t.time()
  print(fibonacci(n))
  times[n]= t.time() - start_time
  print("--- %s seconds ---" % (times[n]))

  start_time = t.time()
  print(fibonacci_DP(n))
  times_DP[n]= t.time() - start_time
  print("--- %s seconds ---" % (times_DP[n]))

def plot(times):
  lists = sorted(times.items())
  x, y = zip(*lists)
  plt.plot(x, y)
  plt.show()
  print(times)

for i in range(50):
  test(i)

plot(times)
plot(times_DP)
```
