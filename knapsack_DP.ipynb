```

items = [310,112,315,223,164,514,343,112,217,414,832]
maxSize = 1363

def knapsack_dp(i,size):
  if(i >= len(items)):
    return 0
  if(size<=0):
    return 0

  if(dp[i][size]!=0):
    return dp[i][size]
  
  _out = knapsack_dp(i+1, size)

  if (size >= items[i]):
    _in = knapsack_dp(i+1, size - items[i]) + items[i]
  
  if(size >= items[i] and _out < _in):
    dp[i][size]=_in
  else:
    dp[i][size]=_out   
  
  print(f"i:{i},size:{size}, dp[i][size]{dp[i][size]}, items[i]:{items[i]}" )
  return dp[i][size]


dp= [[0 for col in range(maxSize+1)] for row in range(len(items)+1)]



value= knapsack_dp(0, maxSize)
print (dp[0][maxSize])
```
